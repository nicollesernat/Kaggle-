```python
pastel_colors = sns.color_palette("pastel", n_colors=10)
```

```python
# Cargar el archivo CSV
df = pd.read_csv("train.csv")

# Contar los 10 departamentos con más estudiantes
departamento_counts = df["E_PRGM_DEPARTAMENTO"].value_counts().head(10)


for departamento, cantidad in departamento_counts.items():
    print(f"- {departamento}: {cantidad}")

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(departamento_counts.index, departamento_counts.values, color=pastel_colors)

# Agregar etiquetas con el número exacto al final de cada barra
for bar, count in zip(bars, departamento_counts.values):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(count),
             ha='center', va='bottom', fontsize=10)

# Etiquetas y título
plt.title("Top 10 departamentos con más estudiantes")
plt.xlabel("Departamento")
plt.ylabel("Cantidad de estudiantes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Cuantos estudiantes hay por cada programa academico
asignatura_counts = df["E_PRGM_ACADEMICO"].value_counts()
asignatura_counts = asignatura_counts.sort_index()

for asignatura, cantidad in asignatura_counts.items():
    print(f"- {asignatura}: {cantidad} estudiantes")
```
